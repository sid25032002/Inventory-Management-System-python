```python
import json
fd = open("example.json",'r')
r = fd.read()
fd.close()
record = json.loads(r)

loop=1
while(loop==1):
    print("\n**********INVENTORY MANAGEMENT SYSTEM**********\n")
    print("Enter 1 to ADD a new item")
    print("Enter 2 to MODIFY an existing item")
    print("Enter 3 to DELETE an existing item")
    print("Enter 4 to VIEW inventory")
    print("Enter 5 to EXIT")
    ch=int(input("\nEnter your choice: "))
    
    if(ch==1):
        prod_id = str(input("\nEnter product id: "))
        if prod_id in record:
            print("Item already exists enter new id")
        else:
            name = str(input("Enter name: "))
            Type = str(input("Enter type: "))
            pr = int(input("Enter price: "))
            qn = int(input("Enter quantity: "))
            dis = int(input("Enter discount: "))
            record[prod_id] = {'name': name, 'type': Type, 'pr': pr, 'qn': qn, 'dis': dis}
            print("\nItem added successfully")
            js = json.dumps(record)
            fd = open("example.json",'w')
            fd.write(js)
            fd.close()
    
    elif(ch==2):
        prod_id = str(input("\nEnter product id: "))
        if prod_id in record:
            name=record[prod_id]['name']
            pr=record[prod_id]['pr']
            Type=record[prod_id]['type']
            dis=record[prod_id]['dis']
            qn = int(input("Enter new stock quantity: "))
            cq=record[prod_id]['qn']+qn
            record[prod_id] = {'name': name, 'type': Type, 'pr': pr, 'qn': cq, 'dis': dis}
            print("\nItem updated successfully\nUpdated quantity: ",record[prod_id]['qn'])
            js = json.dumps(record)
            fd = open("example.json",'w')
            fd.write(js)
            fd.close()
        else:
            print("\nItem dosent exists enter new id")
    
    elif(ch==3):
        prod_id = str(input("\nEnter product id: "))
        if prod_id in record:
            del record[prod_id]
            print("\nItem with prod id: ",prod_id," deleted successfully")
            js = json.dumps(record)
            fd = open("example.json",'w')
            fd.write(js)
            fd.close()
    
    elif(ch==4):
        fd = open("example.json",'r')
        r = fd.read()
        fd.close()
        record = json.loads(r)
        print("\nID\t\tNAME\tTYPE\t\tPRICE\tQUANTITY\tDISCOUNT\n")
        for i in record:
            print(i,"----->",record[i]['name'],"\t",record[i]['type'],"\t",record[i]['pr'],"\t",record[i]['qn'],"\t\t",record[i]['dis'])
    
    else:
        print("\nYou have LOGGED OUT SUCCESSFULLY")
        loop=0
```
